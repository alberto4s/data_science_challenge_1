## Compact solution

```python
filepath = folder_path + filename2
l_destination = ['MAD', 'BCN', 'AGP']

nrows = 100_000
searches = pd.read_csv(filepath,
                      delimiter = '^',
                      nrows = nrows,
                      usecols = ['Date', 'Destination']
                      )

searches['Destination'] = searches['Destination'].str.upper().str.strip()
searches['Date'] = searches['Date'].map(lambda x: str(x))
searches['year-month'] = searches['Date'].str[:7]
destination_fil_cond = searches['Destination'].isin(l_destination)
searches = searches[destination_fil_cond]
grouped = searches.groupby(['Destination', 'year-month']).count().reset_index()
grouped.rename({'Date': 'count'},axis=1,inplace=True)

```

```python
sns.lineplot(data = grouped,
            x = 'year-month',
            y = 'count',
            hue = 'Destination');
```

```python
def clean_data_03(searches, l_destination = ['MAD', 'BCN', 'AGP']):
    searches['Destination'] = searches['Destination'].str.upper().str.strip()
    
    searches['Date'] = searches['Date'].map(lambda x: str(x))
    searches['year-month'] = searches['Date'].str[:7]
    
    destination_fil_cond = searches['Destination'].isin(l_destination)
    searches = searches[destination_fil_cond]
    
    grouped = searches.groupby(['Destination', 'year-month']).count().reset_index()
    grouped.rename({'Date': 'count'},axis=1,inplace=True)
    
    return grouped
```

```python
nrows = 100_000
searches = pd.read_csv(filepath,
                      delimiter = '^',
                      nrows = nrows,
                      usecols = ['Date', 'Destination']
                      )

grouped = clean_data_03(searches)
```

```python
sns.lineplot(data = grouped,
            x = 'year-month',
            y = 'count',
            hue = 'Destination');
```

### Chunks

```python
filepath = folder_path + filename2
chunksize = 1_000_000
nrows = None

iterator_searches = pd.read_csv(filepath,
                                delimiter = '^',
                                nrows = nrows,
                                usecols = ['Date', 'Destination'],
                                chunksize = chunksize
                               )
```

```python
all_grouped = pd.DataFrame()

for i,searches in enumerate(iterator_searches):
    grouped = clean_data_03(searches)
    all_grouped = all_grouped.append(grouped)
    
final_reduced = all_grouped.groupby(['Destination', 'year-month']).sum().reset_index()
final_reduced.rename({'count': 'sum'},axis=1,inplace=True)
```

```python
final_reduced
```

```python
fig, ax = plt.subplots(figsize = (12,8))

sns.lineplot(data = final_reduced,
            x = 'year-month',
            y = 'sum',
            hue = 'Destination',
             ax = ax
            );
```
